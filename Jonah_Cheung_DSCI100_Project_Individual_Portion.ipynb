2. Questions:

Broad question:
   
We would like to know which "kinds" of players are most likely to contribute a large amount of data so that we can target those players in our recruiting efforts.

Specific question:
    
Can a player’s total hours played (played_hours) and age (Age) be used to predict a player's experience level (experience)?

Explanation:

This question will help the researchers identify patterns between player age and experience, which can support more targeted recruitment efforts. In general, players who spend more hours in the game contribute more gameplay data, making them more valuable participants for analysis. Using the players.csv dataset, I plan to create a tibble containing the three relevant columns: Age, played_hours, and experience. I will remove all observations where played_hours equals zero, since these players have not generated gameplay data and are unlikely to contribute meaningful information in the future. From there, I will create a scatter plot of age versus played hours, with player experience represented by colour.



3. Exploratory Data Analysis and Visualization

*note: I have already read both data sets before the first portion

```r
#compute the mean value for all quantitative variables in players.csv
players_mean<-players|>
    select(played_hours, Age)|>
    summarise(
    mean_played_hours = mean(played_hours, na.rm = TRUE),
    mean_age = mean(Age, na.rm = TRUE)
  )
players_mean
```

```r
#Make a few exploratory visualizations of the data to help you understand it

players_scatter_plot<-players|>
    filter(played_hours > 0) |>
    ggplot(aes(x = Age, y = played_hours, colour = experience, shape = experience))+
    geom_point(alpha=0.7)+
    labs(title = "Relationship between Player Age, Played Hours, and Experience", x = "Player Age", y = "Hours Spent on the Server", colour = "Player Experience")
players_scatter_plot
```

*note: Observations with zero played hours were removed prior to calculating the means. These players did not actively contribute gameplay data, so excluding them ensures that the summary statistics reflect only active participants.

observation: The 3 players with the longest play time are all "regulars" regarding experience. Otherwise, there is no obvious trend between these three variables. This ambiguity may be a problem for when I am analyzing this data to solve my question.

```r
#Histogram to observe the age distribution of the data
options(repr.plot.width = 17, repr.plot.height = 6)  

players_age_histogram<-ggplot(players, aes(x = Age)) +
  geom_histogram(binwidth = 1) +
  labs(
    title = "Distribution of Player Ages",
    x = "Age (years)",
    y = "Number of Players"
    )+
    scale_x_continuous(breaks = seq(min(players$Age, na.rm = TRUE), 
                                  max(players$Age, na.rm = TRUE), by = 1))
players_age_histogram
```

observations: the highest concentration of players seem to be from the 17-24 year old range, with 17 being the most common player age.

```r
#Bargraph to observe the gender and subscription distribution of the data
players_gender_bar<-ggplot(players, aes(x = gender, fill = subscribe)) +
    geom_bar() +
    labs(title = "Gender Distribution of Players", x = "Gender", y = "Number of Players") +
    theme(axis.title = element_text(size = 15), 
         axis.text = element_text(size = 7))
players_gender_bar
```

Observations: while there are more male than female subscribers, the ration of subscribed to non-subscribed players are higher in females than males, meaning there is a greater chance for a female to subscribe than a male.

```r
#Bargraph to observe the player experience and subscription distribution of the data
players_experience_bar<-ggplot(players, aes(x = experience, fill = subscribe)) +
    geom_bar() +
    labs(title = "Experience Distribution of Players", x = "Experience level", y = "Number of Players") +
    theme(axis.title = element_text(size = 15), 
         axis.text = element_text(size = 7))
players_experience_bar
```

Observations: The ratios of subscribed to non-subscribed players are about the same across all experience levels.
